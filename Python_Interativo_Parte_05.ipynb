## Tuplas

Para [w3schools](https://www.w3schools.com/python/python_tuples.asp), as tuplas são um tipo de coleção de dados **ordenados e imutáveis**.

As tuplas são **imutáveis**. Portanto, uma vez criadas, elas não podem ser modificadas.

Criar tuplas no Python é muito simples, basta colocar os valores necessários entre **parênteses**. Veja:

````
tupla = (1, 2)
tupla
````

As tuplas são **indexadas**. Logo, você pode acessar o primeiro elemento colocando o número `0` entre colchetes (`[` e `]`).

O **índice** sempre começa em `0`, portanto, `0` corresponde ao primeiro elemento, `1` ao segundo, e assim por diante.

````
tupla[0]
````

````
tupla[1]
````

O acesso com índice negativo é permitido. Acessar com índice igual a `-1` irá retornar o último elemento da tupla.

````
tupla[-1]
````

Como dito, as tuplas são imutáveis, portanto, tentar alterar um valor de uma tupla irá gerar uma excessão:

````
# tupla[1] = 0
````

Consulte o tamanho da tupla com a função **`len`**. O parâmetro da função `len` é a variável que armazena a referência à tupla.

````
len(tupla)
````

Tentar obter o valor de uma tupla além de seu tamanho também irá gerar uma excessão:

````
# tupla[2]
````

As tuplas podem ser formadas por objetos de **diferentes tipos**.

````
tupla2 = (1, '1')
tupla2
````

````
def fun():
    return (1, 2)

fun()
````

Podemos iterar sobre os elementos da tupla

````
tupla = (1, 2, 3, 4)
for elemento in tupla:
    print(elemento)

print()
for i in range(len(tupla)):
    print(tupla[i])
````

## Listas

image.png

Segundo [Geeks for Geeks](https://www.geeksforgeeks.org/python-list/), as listas são como *arrays* em outras linguagens.

As listas podem armazenar **qualquer tipo de valor** e são **mutáveis**. Os elementos da lista são indexados na sequência que são adicionados.

Você pode criar uma lista em Python com o construtor `list()` ou com dois colchetes (`[` e `]`), contendo elementos ou não.

````
lista = list()
````

````
lista = []
````

````
lista = [1, 2, 3, 4]
````

Os elementos de uma lista são indexáveis, então podemos acessar o n-ésimo elemento da lista, da mesma forma como fizemos com as tuplas. Por exemplo:

````
lista[0]
````

Observe que os índices começam em 0 e vão até n - 1, onde n é o tamanho da lista.

Listas são iteráveis, então utilizá-la em um comando `for` é bastante simples.

````
for n in lista:
    # resultado = n**2
    print(n**2, end=' ')
````

Adicione novos elementos ao final com `append`.

````
for i in range(5, 10):
    lista.append(i)
lista
````

Você também pode iterar sobre os índices do vetor, basta iterar sobre o `range` do comprimento total da lista.

O comprimento da lista é obtido com a função `len()`, assim como fizemos com as tuplas.

````
lista2 = []
for i in range(len(lista)):
    lista2.append(lista[i]**2)
lista2
````

O método `pop()` retira o último elemento da lista.

````
print(lista)
print(lista.pop())
print(lista)
````

A função shuffle do módulo random randomiza a lista

````
import random
random.shuffle(lista)
lista
````

O método `sort()` ordena os elementos.

````
print(lista)
lista.sort()
print(lista)
````

## Sets (conjuntos)

Os *sets* são como conjuntos matemáticos. Eles são compostos por valores únicos e desordenados.

Para [Geeks For Geeks](https://www.geeksforgeeks.org/sets-in-python/), o set é um tipo de coleção de dados **não ordenada**, **iterável**, **mutável** e que **não apresenta duplicatas**.

image.png

Criar um set no Python é similar a criar listas e tuplas com o uso do construtor, isto é, chamando o método `set()`.

````
pares = set()
for i in range(0, 10, 2):
    pares.add(i)
pares
````

````
pares.add(88)
pares.add(44)
for numero in pares:
    print(numero)
````

Você pode passar algum objeto iterável como parâmetro para adicionar dados no momento que você criar o objeto.

````
frutas = set(['Abacaxi', 'Melão', 'Laranja'])
frutas
````

````
for fruta in frutas:
    print(fruta)
````

Para adicionar elementos, use o método `add()`.

````
pares.add(2)
pares.add(4) 
pares.add(10) 
pares
````

````
pares.add(2)
pares
````

Observe que a inserção de elementos repetidos não produz nenhum efeito.

Você pode remover elementos utilizando o método `remove()`. O parâmetro é o elemento que você quer remover.

````
if 0 in pares:
    pares.remove(0)
pares
````

## Dicionários

Observe novamente como acessamos os elementos das listas, geralmente utilizamos o índice, certo?

Acontece que o índice das listas são sempre números sequenciais, e que podem ou não ter algum significado interessante.

Os dicionários são como as listas, mas, ao invés de utilizar números como índices, utilizamos **chaves**.

O mais legal é que as chaves podem ser qualquer coisa, desde números, strings, até outros objetos!

image.png

Para [Real Python](https://realpython.com/python-dicts/), os dicionários consistem em uma coleção de pares **chave-valor**, isto é, cada chave (única) é associada a um valor.

*Atenção: as chaves nos dicionários são únicas!*

Criar dicionários no Python também é bastante simples. Para isso, basta abrir as chaves (`{` e `}`) e inserir as chaves e os valores.

Cada par de chave-valor deve ser separado com vírgula, assim como as listas, e deve ser escrito na forma `<chave>: <valor>`. Observe: 

````
capitais = {
     'São Paulo' : 'São Paulo',
     'Paraná'   : 'Curitiba',
     'Mato Grosso': 'Cuiabá',
     'Goiás': 'Goiânia'
}

capitais
````

Acesse o valor por meio dos índices, assim como uma lista:

````
capitais['Goiás']
````

O método `keys()` retorna as chaves do dicionário em uma lista, ao passo que `values()` retorna os valores.

````
capitais.keys()
````

````
capitais.values()
````

Podemos ter dicionários dentro de dicionários.

````
capitais = {
     'Brasil': {
        'São Paulo' : 'São Paulo',
        'Paraná'   : 'Curitiba',
        'Mato Grosso': 'Cuiabá',
        'Goiás': 'Goiânia'
     },
     'Estados Unidos': {
         'Califórnia': 'Sacramento',
         'Flórida': 'Tallahassee'
     }
}

capitais
````

````
capitais['Brasil']['Paraná']
````

### Calculando a proteína a partir do DNA

#### O que é o RNA?

O RNA ou ácido ribonucleico é um acido nucleico, similar ao DNA. Entretanto, diferentemente do DNA que é muito estável, o RNA é um membro muito mais ativo da família de ácidos nucleicos. No contexto da bioinformática, existem duas diferenças muito importantes entre RNA e DNA:

- RNA difere do DNA por um nucleotídeo;
- RNA consiste em um único cordão;
- O RNA é feito das seguintes bases nucleares:
  - Citosina (C);
  - Guanina (G);
  - Adenina (A);
  - Uracil (U).

#### Transcrição de DNA em RNA

Transcrição é o processo onde um cordão de DNA é usado como template para construir um cordão de RNA, por meio da cópia de nucleotídios, onde o uracil é utilizado no lugar da timina.

Considere a seguinte sequência de DNA como exemplo:

`AATTTGCTTGGCAAATCGTATGCCTTGGG`

Você pode obter o cordão RNA relevante como a seguir, substituindo todas as ocorrências de T por U.

`AAUUUGCUUGGCAAAUCGUAUGCCUUGGG`
